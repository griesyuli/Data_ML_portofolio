```python
import requests as rq
from bs4 import BeautifulSoup as bsoup
import pandas as pd
```

# Banyak halaman yang ada

```python
r = rq.get('http://data.jakarta.go.id/dataset?res_format=CSV')
```

```python
r.text [:1000]
```

```python
soup = bsoup(r.text, 'html.parser')
```

```python
r.close()
```

```python
soup
```

```python
pages = []
for pg in soup.findAll('div', attrs={'class':'pagination pagination-centered'}):
    li_tag = pg.findAll('li')
#     print (li_tag)
    for li in li_tag:
        page = li.getText('href')
        pages.append(page)
        
# print (pages)
n_page = int(pages[-2])
print ('Banyak page dalam dataset :', n_page)

    
# <div class="pagination pagination-centered"><ul> 
# <li class="active"><a href="/dataset?res_format=CSV&amp;page=1">1</a></li> 
# <li><a href="/dataset?res_format=CSV&amp;page=2">2</a></li> 
# <li><a href="/dataset?res_format=CSV&amp;page=3">3</a></li> 
# <li class="disabled"><a href="#">...</a></li> 
# <li><a href="/dataset?res_format=CSV&amp;page=159">159</a></li> 
# <li><a href="/dataset?res_format=CSV&amp;page=2">Â»</a></li></ul></div>
```

# Ambil tiap URL data di setiap halaman

```python
url_dataset = []
title_dataset = []

for n in range(n_page):
# for n in range(2):
    n = n+1
    print ('\n ===== Page', str(n))
    url = "http://data.jakarta.go.id/dataset?res_format=CSV&page=%s" % (n)
    print (url)
    r = rq.get(url)
    soup = bsoup(r.text, 'html.parser')
    r.close()
#     print (soup)

    
    for pg in soup.findAll('h3', attrs={'class':'dataset-heading'}):
#         print (pg)
        a_tag = pg.findAll('a')
        for a in a_tag:
            title_data = a.getText('href')
#             print (title_data)
            url_data = a['href']
#             print (url_data)

            title_dataset.append(title_data)
            url_dataset.append(url_data)
           
```

```python
dataset_dict = {'title':title_dataset, 'url':url_dataset}
# print (dataset_dict)
df = pd.DataFrame(dataset_dict, columns= ['title', 'url'])
export_csv = df.to_csv (r'dataset_jakarta.csv', index = None, header=True) 


print (df.head(10))
```
